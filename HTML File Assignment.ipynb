```python
filenames = ['example_html_file.html']
for filename in filenames:
    try:
        with open(filename) as f_obj:
            contents = f_obj.read()
    except FileNotFoundError:  
        msg = "Sorry, the file " + filename + " does not exist."
        print(msg)
    else:
        print("\nDetails of the file (" + filename + ")")
        print("\n******************************************\n")

        # Count approximate number of words in the file.
        words = contents.split()
        num_words = len(words)
        print("Number Of Words: "+str(num_words))

        # Count no. of lines in the file.
        lines = contents.splitlines();
        num_lines = len(lines);
        print("Number Of Lines: " + str(num_lines))

        #Count non-space characters in the file.
        ch = contents
        num_ch = len(ch) - contents.count(' ')
        print("Number Of Non-space Characters: " + str(num_ch))

        #Count spaces in the file.
        spaces = contents.split(' ')
        num_spaces = contents.count(' ')
        print("Number Of Spaces: " + str(num_spaces))
         
        #Count tabs in the file.
        tabs = contents
        num_tabs = tabs.count('    ')
        print("Number Of Tabs: " + str(num_tabs))

        #Count paragraphs in the file.
        paras = contents.splitlines()
        num_para = 0
        for i in paras:
            if i == "" or i == "  ":
                num_para += 1
        print("Number Of Paragraphs: " + str(num_para))

        #HTML Tags used in the file
        num_tags = 0
        tag_started = False
        tags = []
        tag_name = ""

        for c in contents:
            if c == "<":
                tag_started = True
            elif tag_started == True:
                if c == "/" or c == ">" or c == " ":
                    tag_started = False
                    tags.append(tag_name)
                    tag_name=""
                else:
                    tag_name += c   
        tags = [x.strip() for x in tags if x.strip()]
        num_tags = len(tags)
        tags = list(dict.fromkeys(tags))       
        print("HTML Tags Used:",tags)
        print("Number Of HTML Tags: " + str(num_tags))
```
